```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def V_q(r,r_q,q):
    d12 = np.sqrt((r[0]-r_q[0])**2 + (r[1]-r_q[1])**2)
    return q/d12

def V_tot(r,r_qs,qs):
    V = np.zeros_like(r[0])
    for i in range(len(qs)):
        Vqi = V_q(r,np.array(r_qs[i]),qs[i])
        V += Vqi
    return V
```

```python
x,y = np.linspace(-3,3,120) , np.linspace(-3,3,120)
XX , YY = np.meshgrid(x,y)
```

```python
q = 1
r_q = np.array([0,0])
V = V_q([XX,YY],r_q,q)
plt.contourf(XX,YY,V,cmap='Purples')
niveles = np.linspace(0,5,10)
plt.contour(XX,YY,V,levels=niveles,colors='b',linewidths=(1,))
plt.plot(r_q[0],r_q[1],'ro')
plt.gca().set_aspect('equal')
plt.show()
```

```python
qs = [1,1]
r_qs = [[0,1],[0,-1]]
V = V_tot([XX,YY],r_qs,qs)
niveles = np.linspace(0,5,10)
plt.contourf(XX,YY,V,cmap='Purples')
plt.contour(XX,YY,V,levels=niveles,colors='b',linewidths=(1,))
for i in range(len(r_qs)):
    plt.plot(r_qs[i][0],r_qs[i][1],'ro')
plt.gca().set_aspect('equal')
plt.show()
```

```python
qs = [1,-1]
r_qs = [[0,1],[0,-1]]
V = V_tot([XX,YY],r_qs,qs)
plt.contourf(XX,YY,V,cmap='Purples')
niveles = np.linspace(-5,5,10)
plt.contour(XX,YY,V,levels=niveles,colors='b',linewidths=(1,))
plt.plot(r_qs[0][0],r_qs[0][1],'ro')
plt.plot(r_qs[1][0],r_qs[1][1],'go')
plt.gca().set_aspect('equal')
plt.show()
```

```python
qs = [1,1,-1,-1]
r_qs = [[1,1],[-1,-1],[-1,1],[1,-1]]
V = V_tot([XX,YY],r_qs,qs)
plt.contourf(XX,YY,V,cmap='Purples')
niveles = np.linspace(-5,5,10)
plt.contour(XX,YY,V,levels=niveles,colors='b',linewidths=(1,))
for i in range(len(r_qs)):
    if qs[i] > 0:
        plt.plot(r_qs[i][0],r_qs[i][1],'ro')
    else:
        plt.plot(r_qs[i][0],r_qs[i][1],'go')
plt.gca().set_aspect('equal')
plt.show()
```
